```python
import pandas as pd
from pathlib import Path

# Diretórios base
PROJECT_DIR = Path(".").resolve().parents[0] # Isso serve para subir um nível da pasta e chegar na raiz do projeto
NOTEBOOKS_DIR = PROJECT_DIR / "notebooks"
DATA_PROCESSED_OLIST_DIR = PROJECT_DIR / "data" / "processed" / "olist"
DATA_RAW_OLIST_DIR = PROJECT_DIR / "data" / "raw" / "olist"

PROJECT_DIR, NOTEBOOKS_DIR, DATA_PROCESSED_OLIST_DIR, DATA_RAW_OLIST_DIR
```

```python
orders = pd.read_csv(DATA_RAW_OLIST_DIR / "olist_orders_dataset.csv")
customers = pd.read_csv(DATA_RAW_OLIST_DIR / "olist_customers_dataset.csv")
order_items = pd.read_csv(DATA_RAW_OLIST_DIR / "olist_order_items_dataset.csv")
payments = pd.read_csv(DATA_RAW_OLIST_DIR / "olist_order_payments_dataset.csv")
products = pd.read_csv(DATA_RAW_OLIST_DIR / "olist_products_dataset.csv")
```

```python
list_columns_datetime = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in list_columns_datetime:
    orders[col] = pd.to_datetime(orders[col], errors="coerce")
```

```python
orders.info()
```

```python
list_columns_str = [
    "order_id",
    "customer_id",
    "order_status"
]

for col in list_columns_str:
    orders[col] = orders[col].astype(str)
```

```python
orders.info()
```

```python
list_customers_str = [
    "customer_id",
    "customer_unique_id",
    "customer_city",
    "customer_state",
    "customer_zip_code_prefix"
]

for col in list_customers_str:
    customers[col] = customers[col].astype(str)
```

```python
customers.info()
```

```python
list_order_items_str = [
    "order_id",
    "product_id",
    "seller_id"
]

for col in list_order_items_str:
    order_items[col] = order_items[col].astype(str)


list_order_items_float = [
    "price",
    "freight_value"
]

for col in list_order_items_float:
    order_items[col] = order_items[col].astype(float)


order_items["shipping_limit_date"] = pd.to_datetime(order_items["shipping_limit_date"], errors="coerce")
```

```python
order_items.info()
```

```python
list_payments_str = [
    "order_id",
    "payment_type"
]

for col in list_payments_str:
    payments[col] = payments[col].astype(str)


list_payments_int = [
    "payment_sequential",
    "payment_installments"
]

for col in list_payments_int:
    payments[col] = payments[col].astype(int)


payments["payment_value"] = payments["payment_value"].astype(float)
```

```python
payments.info()
```

```python
list_products_str = [
    "product_id",
    "product_category_name"
]

for col in list_products_str:
    products[col] = products[col].astype(str)

list_products_float = [
    "product_weight_g",
    "product_length_cm",
    "product_height_cm",
    "product_width_cm"
]

for col in list_products_float:
    products[col] = products[col].astype(float)
```

```python
products.info()
```

```python
print("ORDERS")
print("linhas:", len(orders))
print("order_id únicos:", orders["order_id"].nunique())
print()

print("CUSTOMERS")
print("linhas:", len(customers))
print("customer_id únicos:", customers["customer_id"].nunique())
print()

print("PRODUCTS")
print("linhas:", len(products))
print("product_id únicos:", products["product_id"].nunique())
```

```python
print("orders -> customers (customer_id)")
mask_orders_customers = orders["customer_id"].isin(customers["customer_id"])
print("  total orders:", len(orders))
print("  customer_id que NÃO estão em customers:", (~mask_orders_customers).sum())
print()

print("order_items -> orders (order_id)")
mask_items_orders = order_items["order_id"].isin(orders["order_id"])
print("  total order_items:", len(order_items))
print("  order_id que NÃO estão em orders:", (~mask_items_orders).sum())
print()

print("payments -> orders (order_id)")
mask_payments_orders = payments["order_id"].isin(orders["order_id"])
print("  total payments:", len(payments))
print("  order_id que NÃO estão em orders:", (~mask_payments_orders).sum())
print()

print("order_items -> products (product_id)")
mask_items_products = order_items["product_id"].isin(products["product_id"])
print("  total order_items:", len(order_items))
print("  product_id que NÃO estão em products:", (~mask_items_products).sum())
```

```python
print("MISSING – ORDERS")
print(orders.isna().sum())
print("-" * 40)

print("MISSING – CUSTOMERS")
print(customers.isna().sum())
print("-" * 40)

print("MISSING – ORDER_ITEMS")
print(order_items.isna().sum())
print("-" * 40)

print("MISSING – PAYMENTS")
print(payments.isna().sum())
print("-" * 40)

print("MISSING – PRODUCTS")
print(products.isna().sum())
```

```python
DATA_PROCESSED_OLIST_DIR.mkdir(parents=True, exist_ok=True)
DATA_PROCESSED_OLIST_DIR
```

```python
tables = {
    "orders": orders,
    "customers": customers,
    "order_items": order_items,
    "payments": payments,
    "products": products,
}

for name, df in tables.items():
    output_path = DATA_PROCESSED_OLIST_DIR / f"{name}.parquet"
    print(f"Salvando {name} em {output_path}")
    df.to_parquet(output_path, index=False)

print("Concluído.")
```
